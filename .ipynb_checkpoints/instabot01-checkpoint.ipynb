## QUESTION 5: LIKE/UNLIKE

### opening profile of dilsefoodie

```python
search_bar=driver.find_element_by_class_name('XTCLo')
search_bar.clear()
search_bar.send_keys('dilsefoodie')
wait=WebDriverWait(driver,20)
dilsefoodie=wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'dilsefoodie')))
dilsefoodie.click()
```

### (1) Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
count=31
num=0
for post in driver.find_elements_by_class_name('_9AhH0'):
    num+=1
    count-=1
    if count==0:
        break
    post.click()
    ## Explicit Wait
    wait=WebDriverWait(driver,25)
    lapl=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
    data=BeautifulSoup(driver.page_source,'lxml')
    if len(data.find(class_='fr66n').find('button').find_all('div'))==2:
        lapl.click()
        print(num,'liked')
    else:
        print(num,'Already liked')
    time.sleep(1)
    driver.find_element_by_class_name('NOTWr').click()
    time.sleep(1)
```

### (2) Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
count=31
num=0
for post in driver.find_elements_by_class_name('_9AhH0'):
    num+=1
    count-=1
    if count==0:
        break
    post.click()
    ## Explicit Wait
    wait=WebDriverWait(driver,25)
    lapl=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
    data=BeautifulSoup(driver.page_source,'lxml')
    if len(data.find(class_='fr66n').find('button').find_all('div'))==1:
        lapl.click()
        print(num,'unliked')
    else:
        print(num,'Already Unliked')
    time.sleep(1)
    driver.find_element_by_class_name('NOTWr').click()
    time.sleep(1)
```

## QUESTION 6: EXTRACTING LIST OF FOLLOWERS

### opening profile of foodtalkindia

```python
search_bar=driver.find_element_by_class_name('XTCLo')
search_bar.clear()
search_bar.send_keys('foodtalkindia')
wait=WebDriverWait(driver,20)
dilsefoodie=wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'foodtalkindia')))
dilsefoodie.click()
```

### 1 Extract the usernames of the first 500 followers of ‘foodtalkindia’

```python
driver.find_element_by_partial_link_text('followers').click()
pop_up=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
limit=500
foodtalkindia_followers=[]
cur=driver.execute_script('return arguments[0].scrollHeight;',pop_up)
while len(driver.find_elements_by_xpath('//div[@class="d7ByH"]//span//a'))<500:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;',pop_up)
    time.sleep(1)
    new_height=driver.execute_script('return arguments[0].scrollHeight;',pop_up)
    print(new_height)
    cur=new_height
for follower in driver.find_elements_by_xpath('//div[@class="d7ByH"]//span//a'):
    if limit==0:
        break
    limit-=1
    print(follower.get_attribute('innerHTML'))
    foodtalkindia_followers.append(follower.get_attribute('innerHTML'))
driver.find_element_by_xpath('//div[@class="WaOAr"]//button//div').click()
```

```python
# length of the foodtalkindia_followers list to verify that we got the exact required number of follower's list
len(foodtalkindia_followers)
```
